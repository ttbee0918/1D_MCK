```python
import numpy as np
import matplotlib.pyplot as plt
# m,c,k
# d,wn
# m,k,d
# var = (m = 1000,c = 4000, k = 60000, x0 = 0, v0 = 1)
def MCK1D(m,c,k,x0,v0):
    t = 0.1*np.array([range(0,1000)])
    # print(t)
    # x0 = 0
    # v0 = 1
    wn = (k/m)**0.5
    print('wn =',wn)
    cc = (4*m*k)**0.5
    d = c/cc
    print('d =',d)
    wd = (1-d**2)**0.5*wn
    print('wd =',wd)
    for t0 in t:
        x = np.exp(-d*wn*t0)*(x0*np.cos(wd*t0)+(v0+d*wn*x0)/wd*np.sin(wd*t0))
    #    print(t0)
    #    print(x)
    plt.plot(t0,x)
def MFD1D(m,freq,d,x0,v0):
    t = 0.1*np.array([range(0,1000)])
    # print(t)
    # x0 = 0
    # v0 = 1
    wn = 2*np.pi*freq
    print('wn =',wn)
    k = m*wn**2
    print('k =',k)
    cc = (4*m*k)**0.5
    c = d*cc
    print('c =',c)
    wd = (1-d**2)**0.5*wn
    print('wd =',wd)
    for t0 in t:
        x = np.exp(-d*wn*t0)*(x0*np.cos(wd*t0)+(v0+d*wn*x0)/wd*np.sin(wd*t0))
    #    print(t0)
    #    print(x)
    plt.plot(t0,x)    
```

```python
plt.show()
plt.xlim((0, 20))  
MCK1D(1000,4000,60000,1,0)
MCK1D(2000,4000,60000,1,0)
MCK1D(4000,4000,60000,1,0)
```

```python
plt.show()
plt.xlim((0, 5))  
MFD1D(1000,10,0.1,1,0)
MFD1D(2000,10,0.1,1,0)
MFD1D(4000,10,0.1,1,0)
```
