```python
import numpy as np
import matplotlib.pyplot as plt
# m: mass
# c: damp
# k: stiffness
# freq: frequency
# d: damping ratio between 0~1
# x0: initial position
# v0: initial velocity

def MFD1D(freq,d,x0,v0):
    t = 0.1*np.array([range(0,1000)])
    wn = 2*np.pi*freq
    wd = (1-d**2)**0.5*wn
    for t0 in t:
        x = np.exp(-d*wn*t0)*(x0*np.cos(wd*t0)+(v0+d*wn*x0)/wd*np.sin(wd*t0))
    plt.plot(t0,x)
    
def MCK1D(m,c,k,x0,v0):
    freq = (k/m)**0.5/2/np.pi
    d = c/(4*m*k)**0.5
    MFD1D(freq,d,x0,v0)
 
```

```python
plt.show()
plt.xlim((0, 20))  
MCK1D(1000,4000,60000,1,0)
MCK1D(2000,4000,60000,1,0)
MCK1D(4000,4000,60000,1,0)
```

```python
plt.show()
plt.xlim((0, 5))  
MFD1D(1,0.1,1,0)
MFD1D(10,0.1,1,0)
MFD1D(100,0.1,1,0)
```
